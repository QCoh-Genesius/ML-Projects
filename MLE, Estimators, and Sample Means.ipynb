Bernoulli

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Bernoulli distribution
p = 0.25
n = 30

# Generate 10,000 samples of size 10 from the Bernoulli distribution
samples = np.random.binomial(1, p, size=(10000, n))

# Calculate the sample mean for each row
sample_means = samples.mean(axis=1)

# Parameters for the normal distribution
mean_normal = p
variance_normal = (p * (1 - p)) / n

# Generate values for the normal distribution
x = np.linspace(0, 1, 1000)
normal_distribution = (1 / (np.sqrt(2 * np.pi * variance_normal))) * np.exp(
    -((x - mean_normal) ** 2) / (2 * variance_normal)
)

def ks(data):
    here = np.sort(data)
    vals = []
    for i in range(len(here)):
        stat = norm.cdf(here[i], loc = p, scale = np.sqrt(variance_normal))
        this = abs(stat-((i+1-.5)/10000)) 
        vals.append(this)
    return max(vals)

# Plot the histogram of sample means and the normal distribution
plt.hist(sample_means, bins=30, density=True, alpha=0.5, label='Sample Means')
plt.plot(x, normal_distribution, 'r', label='Normal Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.legend()
plt.title('Sample Means vs. Normal Distribution')
plt.show()
print(ks(sample_means))
```

Beta

```python
# Parameters for the distribution
p = .1
n = 10

# Generate 10,000 samples of size 10 from the distribution
samples = np.random.beta(p, 1, size=(10000, n))

# Calculate the sample mean for each row
log = np.log(samples)
MLES = 1/(1-log.mean(axis=1))

# Parameters for the normal distribution
mean_normal = p/(p+1)
variance_normal = p*p/(n*(1+p)**4)

# Generate values for the normal distribution
x = np.linspace(0, 1, 1000)
normal_distribution = (1 / (np.sqrt(2 * np.pi * variance_normal))) * np.exp(
    -((x - mean_normal) ** 2) / (2 * variance_normal)
)

def ks(data):
    here = np.sort(data)
    vals = []
    for i in range(len(here)):
        stat = norm.cdf(here[i], loc = mean_normal, scale = np.sqrt(variance_normal))
        this = abs(stat-((i+1-.5)/10000)) 
        vals.append(this)
    return max(vals)

# Plot the histogram of sample means and the normal distribution
plt.hist(MLES, bins=30, density=True, alpha=0.5, label='Sample Means')
plt.plot(x, normal_distribution, 'r', label='Normal Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.legend()
plt.title('Sample Means vs. Normal Distribution')
plt.show()
print(ks(MLES))
```
